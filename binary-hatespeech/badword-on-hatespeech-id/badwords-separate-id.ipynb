```python
import pandas as pd
import torch
from transformers import pipeline

# Check if GPU is available, and set device accordingly
device = 0 if torch.cuda.is_available() else -1

# Load the dataset
df = pd.read_csv('/nas.dbms/fathan/test/multilang-hate-models/data_preprocessed.csv', header = 0)
df.dropna(subset=['text'], inplace=True)

```

```python
import json

# Read sexual-related words from a json file
def load_sexual_words(file_path):
    with open(file_path, 'r') as file:
        words = json.load(file)  # Directly load the list of words
    return words

# Load the sexual words from the JSON file
sexual_words = load_sexual_words("/nas.dbms/fathan/test/multilang-hate-models/binary-hatespeech/misc/dict.json")

# Function to check if text contains any sexual-related word
def contains_sexual_words(text):
    return any(word in text.lower() for word in sexual_words)
```
